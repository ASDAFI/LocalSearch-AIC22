بیایید تا این راه حل را پیاده کنیم! برای راحتی شما تابعی ساخته شده است تا کالا ها را بر حسب ارزش مرتب کند. این تابع $n$ و $V$ را به ترتیب به عنوان ورودی دریافت می‌کند و سپس کالا ها را بر اساس ارزش مرتب می‌کند.
خروجی تابع مجموعه $I$ است.

این مجموعه شامل اعداد $0$ تا $n$ می‌باشد (خود $n$ را شامل نیست.). این مجموعه به ترتیب شماره با ارزش ترین کالا ها را نمایش می‌دهد.

$I[0]$ شماره‌ی با ارزش‌ترین کالا ($i$ آن کالا) در میان کالا ها است.

حالا شما باید راه حل حریصانه خود را پیاده سازی کنید. شما تابعی دارید که به ترتیب $n$ ،  $k$ ، $V$ و $W$ را دریافت می‌کند.

خروجی تابع شما لیستی از یک ها و صفر هاست، که عدد $i$ ام به موجودیت کالای $i$ اشاره دارد.

این تابع را به عنوان نمونه ما تکمیل می‌کنیم اما بدانید که جلوتر باید دست به کد شوید!
    </font>
</div>

```python
def do_greedy_by_values(n, k, V, W):
    # this function does greedy by values
    
    answer = [0] * n
    current_capacity = k
    
    I = sort_by_values(n, V)
    
    for i in I:
        if current_capacity >= W[i]:
            answer[i] = 1
            current_capacity -= W[i]
    
    return answer
```

هر دو راه حل حریصانه ما یک بعد از مساله را حذف می‌کند. مساله ما به هر دو پارامتر وزن و ارزش وابسته است.

ایده‌ای که ممکن است به ذهنتان برسد چگالی است! شما می‌توانید پارامتری جدید به نام چگالی خلق کنید که چگالی کالای $i$ ام به صورت زیر تعریف می‌شود:
$$d_i = \frac{v_i} {w_i}$$

این مقدار چگالی یا همان $\text{density}$ کالای  $i$ ام می‌باشد.

این مقدار بیانگر میزان ارزشی است که به ازای هر یک واحد از وزن کالا کسب می‌کنیم. در این معیار، هر دوی وزن و ارزش دخیل هستند.

و باز هم برای راحتی شما:) تابعی برای مرتب سازی کالا‌ها بر حسب چگالی وجود دارد. خروجی این کالا همانند خروجی مرتب‌سازی بر حسب ارزش می‌باشد.
ورودی های این تابع نیز به ترتیب $n$ ، $V$ ، $W$ می‌باشد.

شما تابعی دارید که به ترتیب $n$ ،  $k$ ، $V$ و $W$ را دریافت می‌کند.

خروجی تابع شما لیستی از یک ها و صفر هاست، که عدد $i$ ام به موجودیت کالای $i$ اشاره دارد.

پیاده‌سازی این رویکرد حریصانه بسیار شبیه رویکرد قبلی است.
