حالا شما باید راه حل حریصانه خود را پیاده سازی کنید. شما تابعی دارید که به ترتیب $n$ ،  $k$ ، $V$ و $W$ را دریافت می‌کند.

خروجی تابع شما لیستی از یک ها و صفر هاست، که عدد $i$ ام به موجودیت کالای $i$ اشاره دارد.

این تابع را به عنوان نمونه ما تکمیل می‌کنیم اما بدانید که جلوتر باید دست به کد شوید!
    </font>

```python
def do_greedy_by_values(n, k, V, W):
    # this function does greedy by values
    
    answer = [0] * n
    current_capacity = k
    
    I = sort_by_values(n, V)
    
    for i in I:
        if current_capacity >= W[i]:
            answer[i] = 1
            current_capacity -= W[i]
    
    return answer
```

حال بیایید تا مساله خود را مدل‌سازی کنیم!)

در این مساله ما $n$ کالا  (شی) داریم که هر کالا ارزش $v_i$ را دارد. می‌دانیم که هر کالا در کوله‌پشتی وجود دارد یا وجود ندارد که می‌توانیم این ویژگی را با $x_i$ به نمایش بگذاریم. این مقدار همواره $0$ یا ‌$1$ می‌باشد. این امر را می‌توانیم به شکل زیر بیان کنیم:
$$x_{i}\in \{0,1\}$$

می‌دانیم که ارزش ذاتی و وزن ذاتی هر کالا، همان وزن و ارزش آن است که به خود کالا برمی‌گردد. انتخاب کردن یا نکردن آن کالا تاثیری در ارزش و وزن ذاتی آن ندارد.


حالا بیایید تا مفهوم ارزش حقیقی و وزن حقیقی هر کالا را برای خود تعریف کنیم.
ما می‌دانیم که اگر کالایی در کوله‌پشتی نباشد، ارزش و وزنی برای ما نخواهد داشت. در نتیجه نه هزینه (وزن) و نه سود (ارزش) ای از آن ما را شامل می‌شود.
ارزش حقیقی و وزن حقیقی کالاهایی که در کوله‌پشتی نیستند را می‌توانیم $0$ در نظر بگیریم.

می‌دانیم ارزش حقیقی ای که هر کالا برای ما دارد $x_i \times v_i$ می‌باشد. چرا؟ به این علت که اگر کالا در کوله‌پشتی باشد $x_i$ برابر با $1$ است در نتیجه  $x_i \times v_i$ برابر $v_i$ می‌شود. پس این کالا برای ما $v_i$ ارزش دارد. حالا اگر این کالا در کوله‌پشتی نباشد، $x_i$ برابر با $0$ است در نتیجه  $x_i \times v_i$ برابر $0$ می‌شود. پس این کالا برای ما هیچ ارزشی ندارد. مهم نیست که ارزش ذاتی کالا چقدر است، اگر کالا درون کوله‌پشتی نباشد هیچ ارزشی برای ما ندارد (ارزش حقیقی آن $0$ است.).

با این اوصاف می‌توانیم بیان کنیم که ارزش کوله‌پشتی جمع‌آوری شده، برابر مجموع ارزش های حقیقی است. زیر در نظر گرفتیم که اگر کالا در کوله‌پشتی نباشد ارزش ذاتی آن بی‌اهمیت است و ارزش حقیقی آن $0$ می‌باشد. اعلام کردیم که اگر کالا درون کوله‌پشتی باشد ارزش حقیقی آن $v_i$ می‌شود.

همین موضوع برای وزن نیز صدق می‌کند. وزن کوله‌پشتی جمع‌آوری شده توسط ما، بنا به استدلال قبلی این بار برای وزن،‌برابر مجموع وزن حقیقی تمام کالا هاست.

در نتیجه می‌توانیم وزن و ارزش کوله‌پشتی جمع‌آوری شده را به صورت زیر بنویسیم:

$$ \text{Total value}= \sum_{i=1}^{n}v_i \times x_i$$
$$ \text{Total weight} = \sum_{i=1}^{n}w_i \times x_i$$

هدفی که ما داریم بیشینه کردن ارزش کوله‌پشتی است. همزمان باید حواسمان به وزن کوله‌پشتی جمع آوری شده باشد تا از محدودیت گذر نکند. می‌توانیم مدل مساله را به صورت زیر بنویسیم:
        
$$ \text{maximize}  \sum_{i=1}^{n}v_i \times x_i  \ \ \ \ \ \ \text{s.t} \ \ \ \ \ \ \  \sum_{i=1}^{n}w_i \times x_i \leqslant k$$
$$x_{i}\in \{0,1\}$$

این عبارت بیانگر این مورد است که ما می‌خواهیم ارزش کوله‌پشتی را بیشینه کنیم به نحوی که ($\text{s.t}$) وزن کوله‌پشتی جمع‌شده از محدودیت وزن کوله‌پشتی گذر نکند.

ما به دنبال $X$ ای هستیم که $\lvert X \rvert = n$ (تعداد اعضای آن برابر n باشد.) و $x_i \in X \Rightarrow x_i \in \{0,1\}$ (همه اعضای آن صفر یا یک باشند.).

$X$ همان جواب مساله ما و خروجی تابع ما می‌باشد.

ساده‌ترین ایده‌ای که برای حل این مساله می‌تواند به ذهن همه برسد، بررسی همه حالات است. در این راه حل ما همه جواب های ممکن را بررسی می‌کنیم و بهترین جواب را خروجی می‌دهیم.

این راه حل مانند روش حریصانه نخواهد بود و به وضوح بهترین جواب را به ما خروجی خواهد داد.

می‌دانیم به ازای هر $X$ با شرط‌های ذکر شده، یک جواب برای مساله خود داریم. ما باید تمام $X$ های موجود را بررسی کنیم و ببینیم که کدام موردِ ممکن،  دارای بیشترین ارزش است.

می‌دانیم که $X$ دارای $n$ عضو است که هر عضو آن را $0$ یا $1$ تشکیل می‌دهد. در نتیجه هر عضو $X$ دارای دو حالت است پس (بنا به اصل ضرب!) ما $2^n$ مقدار $X$ متمایز داریم.

حال برای بررسی همه این حالات باید ایده‌ای داشته باشیم. باید بتوانیم حالت‌های جدید را تولید کنیم.

یکی از ایده های ساده‌ای که وجود دارد، کمک گرفتن از اعداد در مبنای دو است. اگر جواب $X$ را داشته باشیم که $X = \{x_1, x_2, x_3, \dots\, x_n\}$ می‌توانیم عدد $B$ را این‌گونه تعریف کنیم:
$$B = x_1 x_2 x_3 \dots x_n$$
عدد $B$ شامل ارقام 0 و 1 است، در نتیجه این عدد در مبنای دو می‌باشد.

می‌توانیم نتیجه بگیریم که به ازای هر $X$ یک عدد $B$ که در مبنای دو است وجود دارد.

می‌دانیم که $B$ دارای $n$ رقم است، در نتیجه در تناظر هر $B$ (عدد $n$ رقمی در مبنای دو) یک عدد صحیح $q$ بین 0 تا $2^n$ وجود دارد.

چرا این گزاره صحیح است؟ زیرا می‌توانیم $B$ را به مبنای ده ببریم. اگر $B$ را به مبنای ده ببریم می‌تواند حداقل 0 و حداکثر $ 2^n - 1$ باشد.


کافیست اعداد صحیح بین  $0$ و $2 ^ n$ را به مبنای دو ببریم تا هر یک بیانگر یک حالت جواب($X$) باشد.

یا به عبارتی روند زیر را برای رسیدن به یک جواب $X$ طی کنیم:
$$ q \ \ \ \rightarrow \ \ \ B \ \ \ \rightarrow \ \ \ X$$ 
        
به طور مثال نمونه زیر برای $n=3$ یا همان زمانی است که $3$ کالا داریم:
$$
0  \rightarrow \ \ \   000   \rightarrow \ \ \  \{0, 0, 0\} \\
1  \rightarrow \ \ \   001   \rightarrow \ \ \  \{0, 0, 1\} \\
2  \rightarrow \ \ \   010   \rightarrow \ \ \  \{0, 1, 0\} \\
3  \rightarrow \ \ \   011   \rightarrow \ \ \  \{0, 1, 1\} \\
4  \rightarrow \ \ \   100   \rightarrow \ \ \  \{1, 0, 0\} \\
5  \rightarrow \ \ \   101   \rightarrow \ \ \  \{1, 0, 1\} \\
6  \rightarrow \ \ \   110   \rightarrow \ \ \  \{1, 1, 0\} \\
7  \rightarrow \ \ \   111   \rightarrow \ \ \  \{1, 1, 1\} 
$$
با توجه به این توضیحات، تابعی برای حل سوال با بررسی همه حالات می‌نویسیم:
    </font>
</div>

```python
def do_bruteforce(n, k, V, W):
    # Do bruteforce
    
    # initialize answer
    max_total_value = 0
    result = []
    
    # check for i(q) in [0, 2^n)
    for i in range(2 ** n):
        X = []
        X_length = 0
        
        # base 10 to binary
        while i != 0:
            x = i % 2
            i //= 2
            X.append(x)
            X_length += 1
            
        X += [0] * (n - X_length)
        X.reverse()  ###  Extra Computation But make it more clear
        
        #print(X)  # view X 
        
        # Evaluate our answer(X)
        total_value = 0
        total_weight = 0
        
        for j in range(n):
            total_weight += W[j]*X[j]
            total_value += V[j]*X[j]
        
        # Check if we find better answer(X)
        if total_value > max_total_value and total_weight <= k:
            max_total_value = total_value
            result = X[:]
    
    return result
```

قبل‌تر گفتیم که ما باید به دنبال جواب مناسبی باشیم. اما مناسب به چه معناست؟ ما باید بتوانیم این عبارت را تعریف کنیم و بتوانیم میان چند جواب مناسب‌ترین را انتخاب کنیم. برای بررسی مناسب بودن یک پاسخ، باید به دنبال یک روند ارزیابی باشیم.

ما باید تابع ارزیابی‌ای داشته باشیم که میزان خوب یا بد بودن جوابی که به‌دست آوردیم (کوله‌پشتی‌ای که یافتیم) را نشان دهد. در مساله‌های مختلف ما این‌کار را با تابع خطا (loss) یا تابع ارزش (value) تعریف می‌کنیم. اگر معیار ارزیابی ما ارزش باشد، هدف ما یافتن پاسخی است که مقدار ارزش آن بیشینه باشد. اما اگر معیار ما در مساله خطا باشد باید پاسخی را بیابیم که مقدار خطا برای آن کمینه است.

مساله کوله‌پشتی ما با تابع ارزش سنجیده می‌شود. اما در مساله‌ای همچون $n$ وزیر($n$ وزیر را در صفحه شطرنج $n \times n$ که هم را تهدید نکنند.)، می‌توانیم تابعی همچون تابع خطا تعریف کنیم که بیانگر تعداد وزیر هایی است که هم را تهدید می‌کنند. هدف ما در این مساله، یافتن پاسخی است که تابع خطا برای آن کمینه باشد.

البته برای این مساله تابع ارزش هم قابل دریافت است. احتمالا برایتان سوال باشد که چگونه؟ 

می‌دانیم که برای ایجاد شرایط تهدید، به دو وزیر نیازمندیم. اگر به ازای هر دو وزیری که در صفحه انتخاب می‌کنیم، هر دو وزیر همدیگر را تهدید کنند، در این شرایط بیشینه مقدار برای تهدید وجود دارد. در نتیجه می‌توانیم بگوییم که بیشینه تهدید $\binom{n}{2}$ است.

حال می‌توانیم تابع ارزش خود را این‌گونه تعریف کنیم:
$$\text{value}(S) = \binom{n}{2} - \text{threats}$$

در اینجا $\text{threats}$ تعداد کل تهدید‌ها در صفحه شطرنج(در پاسخ $S$) و $S$ نیز جوابی که ما یافتیم می‌باشد(یک شیوه چینش $n$ وزیر در صفحه شطرنج). 

می‌دانیم بیشینه مقدار تهدید ها $\binom{n}{2}$ است، در نتیجه مقدار این تابع همواره بزرگتر مساوی $0$ است. هر چه این تابع مقدار بزرگتری را به ما برگرداند، نشان کمتر بودن تهدید ها و مناسب تر بودن جواب است.

دیدید که بسیار ساده توانستیم هر دو تابع ارزش و خطا را برای این مساله تعریف کنیم. انتخاب این‌که مساله را بر اساس کدام تابع حل کنیم، بر عهده ماست و به مدل سازی ما بستگی دارد. ما در مدل سازی کوله‌پشتی از تابع ارزش استفاده کردیم.

تابع ارزش ما در این مساله کوله‌پشتی بدین گونه تعریف شد:
$$ \text{value(X)}= \sum_{i=1}^{n}v_i \times x_i$$
$$x_{i}\in \{0,1\}$$

در اینجا $X$ یک جواب معتبر برای مساله ما می‌باشد.

فرض کنید ما مسیری را رفته‌ایم و این‌که این مسیر به کمک چه راه حلی بوده برای ما اهمیتی ندارد. این مسیر می‌تواند یک ایده‌ی حریصانه‌‌ی ساده باشد یا می‌تواند یک الگوریتم بهینه‌سازی پیچیده‌ باشد.
فی‌الحال چیزی که برای ما مهم است، بودن ما در این نقطه است.
ما می‌توانیم از این نقطه به جهات مختلفی حرکت کنیم. حرکت به سمت هر جهت، جواب متفاوتی به ما می‌دهد. اگر ما یک قدم در هر جهت حرکت کنیم، به یک جواب بسیار شبیه به جواب فعلی خود می‌رسیم.
حال ما باید تمام نقاط اطراف خود(نقاط در همسایگی راه حل فعلی)  را دریابیم و مناسب‌ترین را پیدا کنیم.

حال چگونه می‌توانیم این نقاط را پیدا کنیم؟ ما باید قدم های خود را بر اساس تعدادی عملیات ریاضی تعریف کنیم.
عملیات چیست؟ عملیات تابعی است که اگر پاسخ $X$ را به آن ورودی دهیم، پاسخ جدیدی به ما خروجی خواهد داد.

فرض کنید عملیات $p$ را برای مساله کوله‌پشتی بدین نحو تعریف کنیم: در این عملیات مولفه $0$ ام $X$ یعنی $x_0$ معکوس می‌شود(اگر $0$ بود $1$ و اگر ‌$1$ بود، $0$ می‌شود.).

برای نمونه اگر $X = \{0, 1, 0\}$ باشد، $p(x) = \{1, 1, 0\}$ خواهد شد.
دیدیم که این تابع به ازای دریافت پاسخ $X$ یک پاسخ متفاوت را به ما برمی‌گرداند.

تابع $p$ یکی از قدم های ما است. ما قدم‌های متعددی در این نقطه می‌توانیم تعریف کنیم. می‌توانیم دید خود را باز‌تر کنیم و انتخاب های بیشتری برای خود بسازیم. برای ساخت هر قدم نیاز به یک تابع است.

```python
def select_best_neighbour(n, k, neighbours, V, W):
    best_neighbour = []
    best_neighbour_value = 0
    
    for neighbour in neighbours:
        neighbour_weight = get_weight(n, neighbour, W)
        
        if(neighbour_weight <= k): # check for valid neighbours
            neighbour_value = get_value(n, neighbour, V)
            if(neighbour_value > best_neighbour_value):  # check for better neighbours
                best_neighbour = neighbour
                best_neighbour_value = neighbour_value
    
    return best_neighbour
                
            
    
```

```python
def do_local_search(n, k, V, W):
    answer = [0] * n  # initial first answer
    current_value = 0
    while True:
        neighbours = get_all_neighbours(n, answer)
        selected_neighbour = select_best_neighbour(n, k, neighbours, V, W) # check for best neighbour among neighbours
        selected_neighbour_value = get_value(n, selected_neighbour, V)  
        
        if selected_neighbour_value > current_value:  # check for better neighbour
            answer = selected_neighbour
            current_value = selected_neighbour_value
        else:
            # there is no better neighbour here
            break
    
    return answer
```

بیایید تا با مفهومی به نام اتصال آشنا شویم. فرض کنید که مجموعه $P$ همان مجموعه توابعی است که به کمک آن همسایگی های یک پاسخ را تعریف می‌کردیم. 

اگر بتوانیم به کمک توابع درون مجموعه $P$ از یک جواب $A$ پس از طی کردن تعدادی عملیات که درون $P$ هستند به $B$  برسیم، می‌توانیم بگوییم که جواب $A$ به جواب $B$ پیوسته است.

می‌توانیم این مورد را به صورت زیر بیان کنیم:
$$A = S_1 \rightarrow S_2 \rightarrow S3 \rightarrow  \dots \rightarrow  S_n = B$$
هر کدام از این فلش ها، عملیاتی هستند که در مجموعه $P$ ما وجود دارند.

اگر بخواهیم کمی شهودی تر ماجرا را بررسی کنیم، متوجه می‌شویم که اگر بتوان با قدم زدن از طریق همسایه ها در فضای مسئله، از پاسخ $A$ به پاسخ $B$ رسید، آنگاه پاسخ $A$ به پاسخ $B$ متصل است.

حال سوالی وجود دارد، اگر $A$ به $B$ متصل باشد، آیا لزوما $B$ به $A$ نیز متصل است؟
پاسخ این سوال خیر است! زیرا ممکن است از طریق توابع درون مجموعه $P$ نتوان از $B$ به $A$  رسید. 

از آنجایی که تعریف مجموعه $P$ و توابع درون آن به عهده ماست، ساخت مثال نقض برای این مورد بسیار آسان می‌باشد.

همچنین می‌دانیم که رابطه اتصال متعدی است. بدین معنا که اگر $A$ به $B$ متصل باشد و $B$ به $C$ متصل باشد، آن‌گاه $A$ به $C$ متصل است.

زیرا ما می‌توانیم از $A$  به $B$ با دسته ای از عملیات های درون $P$ دست پیدا کنیم، از $B$ به $C$ نیز می‌توانیم با دسته‌ای دیگر از عملیات های درون $P$ دست پیدا کنیم. درنتیجه می‌توان از $A$ به $C$ را با دسته‌ای از عملیات های $P$ طی کرد.


فرض کنید ما مجموعه $P$ را با توابعی تشکیل دادیم که اگر $A$ به $B$ متصل باشد، $B$ نیز به $A$ متصل است. (رابطه اتصال دوسویی است.)

حال چه می‌شود اگر دو جواب به خصوص در مساله داشته باشیم که به هم متصل نباشند؟

فرض کنیم که جواب های $A$ و $B$ به هم متصل نیستند. همچنین می‌دانیم جواب $X$ در مساله وجود دارد که بهترین پاسخ است.

حال بیایید حالت‌بندی‌ای را انجام دهیم:


1- $X$ به $A$ متصل باشد:
اگر $A$ به بهترین جواب متصل باشد، می‌توان نتیجه گرفت که $B$ به بهترین جواب متصل نیست. زیرا اگر $B$ به $X$ متصل باشد، به دلیل این که $X$ به $A$ متصل است، $A$ با $B$ متصل می‌شود که خلاف فرض ما است.

حال اگر الگوریتم بهینه‌سازی خود را از راه $B$ آغاز کنیم، به هیچ عنوان نمی‌توانیم به بهترین جواب، یعنی $X$ برسیم.


2- $X$ به $B$ متصل باشد:
اگر $B$ به بهترین جواب متصل باشد، می‌توان نتیجه گرفت که $A$ به بهترین جواب متصل نیست. زیرا اگر $A$ به $X$ متصل باشد، به دلیل این که $X$ به $B$ متصل است، $B$ با $A$ متصل می‌شود که خلاف فرض ما است.

حال اگر الگوریتم بهینه‌سازی خود را از راه $A$ آغاز کنیم، به هیچ عنوان نمی‌توانیم به بهترین جواب، یعنی $X$ برسیم.


3- هیچ ‌کدام از ‌$A$ و ‌$B$ به $X$ متصل نباشند:
در این حالت نمی‌توانیم با شروع از $A$ یا $B$ به بهترین جواب مساله برسیم.


4- هر دوی $A$ و $B$ به $X$ متصل باشند:
اگر حالت ما این باشد، می‌توانیم نتیجه بگیریم که $A$ به  $B$ متصل است، در نتیجه خلاف فرض اولیه ما است و این حالت وجود ندارد.


دیدیم در این حالت اگر دو جواب نامتصل وجود داشته باشد، احتمال آن که به بهترین پاسخ نرسیم وجود دارد.

برای نجات از local minima یا همان مینیموم موضعی چه ایده‌ای دارید؟ آیا راهی وجود دارد که از این مشکل فرار کرد؟

ایده‌ای که وجود دارد استفاده از رویکرد شانسی است! تمام همسایه های اطراف ما باید شانس بازدید شدن داشته باشند. اما این شانس برابر است؟ اگر این‌گونه باشد ما به طور کاملا تصادفی و غیر هوشمندانه در حال گشت در محیط مساله هستیم.

می‌توانیم شانس را برای همسایه‌های بهتر بالاتر در نظر بگیریم. با این‌کار رویکرد ما بسیار هوشمندانه تر می‌شود. این حرکت شانسی، برای ما مهلت خروج از local minima و شرایطی که در آن گیر کرده‌ایم را ایجاد می‌کند.

حالا سوال اینجاست که تا کی باید به این روند در الگوریتم خود ادامه دهیم؟ هر چقدر که بخواهیم! هر چه بیشتر در محیط پاسخ‌ها جستجو کنیم، شانس یافتن پاسخ بهتر بیشتر است.
        
فرض کنید می‌خواهیم از بین چند جواب که هر کدام ارزش مشخصی دارند، یک جواب را به صورت شانسی انتخاب کنیم، به نحوی که شانس جواب‌های بهتر به نسبت بهتر بودنشان بیشتر باشد.
        
اول از همه به همه‌ی ارزش‌ها مقدار $1$ را اضافی می‌کنیم تا ارزشی معادل $0$ در میان آن‌ها نباشد و همه شانس انتخاب شدن را داشته باشند. حالا اگر مجموع تمام جواب‌های داخل همسایگی را حساب کنیم، و ارزش هر پاسخ را بر آن تقسیم کنیم، ارزش های ما مقادیری بین $0$ و $1$ پیدا خواهند کرد. می‌توان گفت که این ارزش‌های جدید، شانس انتخاب شدن پاسخ‌ها هستند.
        
تابع ما مجموعه ارزش پاسخ‌ها را دریافت می‌کند و شانس انتخاب هر کدام را باز میگرداند. 
        
حال این تابع را تشکیل می‌دهیم:
    </font>
</div>

```python
def get_chances(values):
    length = len(values)
    
    sum_of_values = 0
    chances = []
    
    for i in range(length):
        chances.append(values[i] + 1)  # to avoid zero chance!
    
    sum_of_values = sum(chances)
    
    for i in range(length):
        chances[i] /= sum_of_values  # calculate chance
    
    return chances
        
```

حالا تابعی می‌خواهیم تا با دریافت این شانس‌ها، موردی را برای ما انتخاب کند. در اینجا روشی را به نام roulette wheel به شما معرفی می‌کنیم که این امر را برای شما بسیار ساده می‌کند.
        
<img src='images/roulettewheel-fun.jpg' style="width:50%;height:50%;">
 

نیاز است تا جمع تجمعی (cumulative sum) را برای شانس های خود تعریف کنیم. در این جمع هر خانه از آرایه  شانس ها برابر مجموع تمام خانه های قبل خود می‌شود.
به طور مثال اگر آرایه شانس های ما برابر $[0.1, 0.5, 0.4]$ باشد، جمع تجمعی آن برابر $[0.1, 0.6, 1]$ می‌شود.

حال یک عدد تصادفی میان $0$ و $1$ تولید می‌کنیم. و به از عضو اول آرایه جمع تجمعی آن شروع به بررسی می‌کنیم.
اولین عضوی که به آن می‌رسیم و عدد ما کوچکتر مساوی آن است، پاسخ مساله است!

<img src='images/roulettewheel.jpg' style="width:70%;height:70%;">
 


حال بیایید تا تابع مورد نظر را تشکیل دهیم. تابع ما آرایه شانس ها را تعریف می‌کند و یک عدد صحیح خروجی می‌دهد که بین 0 و تعداد اعضای آرایه شانس (که بیانگر تعداد همسایه ها می‌باشد!) است.
اگر این عدد صحیح $t$ باشد، بیانگر این است که، $t$ امین همسایه انتخاب شده است.
    </font>
</div>

```python
import random

def  do_roulette_wheel(chances):
    length = len(chances)
    
    comulative_sum = [chances[0]]
    for i in range(1, length):
        comulative_sum.append(chances[i] + comulative_sum[-1])
    
    r = random.random()
    
    for i in range(length):
        if(comulative_sum[i] >= r):
            return i
    
```

```python
def select_neighbour(n, k, neighbours, V, W):
    values = []
    
    for neighbour in neighbours:
        neighbour_weight = get_weight(n, neighbour, W)
        
        if(neighbour_weight <= k): # check for valid neighbours
            neighbour_value = get_value(n, neighbour, V)
            values.append(neighbour_value)
        else:
            values.append(-1)  # 0 chance for infeasible neighbours
            
    chances = get_chances(values)
    index = do_roulette_wheel(chances)
    
    selected_neighbour = neighbours[index]
    
    return selected_neighbour
```

حال باید تابع حل مساله را بنویسیم. در این تابع از روش جدید برای انتخاب همسایه استفاده می‌شود:)

در تابع جدید ما ورودی جدید $\text{iterations}$ وجود دارد که تعداد دفعات تکرار الگوریتم را بیان می‌کند.

هر چه این تعداد بیشتر باشد، شانس یافتن جواب صحیح بیشتر است.

```python
def do_better_local_search(n, k, V, W, iterations):
    best_answer = [0] * n  # initial first answer
    max_value = 0
    
    current_state = best_answer[:]   # initial first state
    
    for i in range(iterations):
        neighbours = get_all_neighbours(n, current_state)
        selected_neighbour = select_neighbour(n, k, neighbours, V, W) # check for some good neighbour
        selected_neighbour_value = get_value(n, selected_neighbour, V)  
        
        if selected_neighbour_value > max_value:  # check for better neighbour
            best_answer = selected_neighbour
            max_value = selected_neighbour_value
        
        current_state = selected_neighbour    # change current state
    
    return best_answer
        
    
```
