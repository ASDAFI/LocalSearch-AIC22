جوابی که می‌توان داد به وضوح خیر است! در بسیاری از مسئله‌ها، فضای مسئله بسیار بزرگ است، بدین معنا که بررسی تمامِ آن هزینه بسیار زیادی دارد. هزینه‌ای اعم از زمان، حافظه و … که واقعا کاستن آن دارای پیچیدگیست.

در دنیای امروز که ما با داده‌های کلان رو در رو هستیم، تعداد حالت‌های مسئله ما بسیار زیاد است، به قدری که بررسی همه حالت‌ها، یا حتی بیشتر آن آرزویی بیش نیست:)

چالشی که امروز علوم کامپیوتر را درگیر کرده است، حل همین مسئله‌ها می‌باشد. اگر نمی‌شود تمام حالات را بررسی کرد، راهی وجود دارد که بدون بررسی همه یا بیشترِ حالات به بهترین جواب رسید؟

            حریصانه        
        </font>
        <p></p>
مشخص است که اگر از شما بخواهند میان تعدادی جایزه یک جایزه را انتخاب کنید، با ارزش‌ترین آن ها را برای خود انتخاب می‌کنید.

ایده‌ای که می‌بینید، یکی از ساده ترین رویکرد‌ها در حل مسئله‌ است.

در این ایده شما کاری را می‌کنید که در این لحظه بیشترین سود را ببرید.

مثلا وقتی تعدادی اسکناس روی زمین است و به شما می‌گویند که می‌توانید ۱۰ اسکناس به انتخاب خود از روی زمین بردارید، شما در هر مرحله با ارزش‌ترین اسکناس را برخواهید داشت.

با این روش، شما حریصانه عمل خواهید کرد و در هر مرحله انتخابی می‌کنید که بیشترین سود لحظه‌ای را ببرید.

باید دقت کنید که بیشترین سود در هر لحظه، لزوما بیشترین سود را در آخر به شما نخواهد داد مگر اینکه این موضوع را بتوانید ثابت کنید.

بیایید تا راه حل حریصانه خود را برای مسئله اسکناس‌ها بررسی کنیم.

از برهان خلف استفاده می‌کنیم. فرض کنیم که از این روش استفاده کردیم و به ترتیب با ارزش ترین اسکناس‌ها را از روی زمین برداشتیم و طبق فرض این بهترین حالت نیست و حالتی وجود دارد که اسکناس‌های جمع‌آوری شده با ارزش تر است.
می‌دانیم که این دو جواب متفاوت هستند، در نتیجه در تعدادی از اسکناس ها تفاوت دارند، بدین معنا که ما برای این‌که به بهترین جواب برسیم باید تعدادی اسکناس را بر روی زمین قرار دهیم و تعدادی اسکناس جدید از روی زمین برداریم.
ما می‌دانیم که هر اسکناسی که اکنون در دست داریم با ارزش تر از همه اسکناس های روی زمین است یا نهایتا با ارزش ترین 
اسکناس روی زمین هم ارزش است.

در نتیجه اگر یک اسکناس را از دست خود به روی زمین بگذاریم و اسکناسی جدید برداریم، حداکثر ۰ واحد به ارزش اسکناس های درون دست ما اضافه خواهد شد. حال اگر $n$ اسکناس را از دست خود روی زمین قرار دهیم و $n$ اسکناس دیگر را از روی زمین برداریم، حداکثر $0 \times n$ واحد یعنی ۰ واحد به ارزش اسکناس های درون دست ما اضافه خواهد شد.
در نتیجه راه حل جدید که راه حلی متفاوت است، از راه حل ما بهتر نیست. تناقضی که در اینجا به آن برخورد می‌کنیم این موضوع را ثابت می‌کند.

در اینجا ما راه حل حریصانه را برای اسکناس ها بررسی کردیم و ثابت کردیم این الگوریتم در هر مرحله علاوه بر انتخاب بهترین گزینه لحظه‌ای، بهترین گزینه را برای کل مسئله انتخاب می‌کند.

            کمی کد!        
        </font>
        <p></p>
حال بیایید برای دست‌گرمی کمی تمرین کنیم.

ما در اینجا pseudo code الگوریتم حریصانه را بررسی خواهیم کرد و تمرینی بر آن خواهیم داشت.

pseudo code (سودو کد خوانده می‌شود!) زبانیست متفاوت از تمام زبان‌های برنامه‌نویسی که درک و فهم آن برای انسان بسیار ساده‌تر است.
برنامه‌نویسی به کمک سودو کد ها بسیار راحت می‌باشد زیرا بسیار خوانا است و درک آن ها نیاز به علم به زبان برنامه‌نویسی خاصی ندارد. می‌توان گفت سودو کد زبان مشترک میان تمام برنامه‌نویس هاست.

با سودو کد الگوریتم حریصانه آغاز می‌کنیم:

<img src='images/greedy-pseudo code.png' style="width:50%;height:50%;">


بیایید تا خط به خط جلو برویم:

1- ما باید مجموعه ای از انتخاب ها داشته باشیم که جواب مساله ما است.
 ما مجموعه جواب $S$ را برابر تهی قرار می‌دهیم.
 
2- در مرحله بعد به ازای تمام انتخاب‌ها، هزینه انتخاب را حساب می‌کنیم. اصطلاح هزینه همیشه اصطلاحی منفی بوده است و ما به دنبال کمینه کردن مقدار آن هستیم. در مساله ما بحث ارزش وجود دارد که ما می‌خواهیم آن را بیشینه کنیم.

3- حالا تا زمانی که $S$ یک راه حل ممکن و کامل نشده است باید کار‌هایی را انجام دهیم. منظور از ممکن و کامل بودن این است که $S$ به عنوان یک جواب مساله قلمداد شود. در مورد مساله اسکناس ها می‌توانیم بگوییم که $S$ باید حاوی $n$ عضو باشد زیرا ما باید $n$ اسکناس انتخاب کنیم.

4- حال در هر مرحله از اجرای این حلقه ما باید انتخاب $s$ را از مجموعه انتخاب های $E$ پیدا کنیم که کمترین هزینه را برای ما دارد. در مساله اسکناس ها ما باید مساله را برای ارزش ها حل کنیم درنتیجه باید انتخابی را برگزینیم که بیشترین ارزش را داشته باشد. بیشترین ارزش معادل با کم‌ترین هزینه می‌تواند باشد.

5- حال انتخاب خود $s$ را به مجموعه انتخاب ها $S$ اضافه می‌کنیم.

6- حالا باید $s$ را از مجموعه همه انتخاب‌ها یعنی $E$ حذف کنیم. در مرحله بعد نیز مقدار هزینه جدید (یا ارزش جدید) را برای هر انتخاب $e$ که در $E$ واقع است می‌یابیم.
در آخر می‌دانیم که $S$ پاسخ حریصانه ما به مساله است.


اگر بخواهیم الگوریتم حریصانه را جمع‌بندی کنیم، شما باید در هر مرحله حریصانه عمل کنید و فقط به فکر این لحظه خود ( یا لحظات محدودی) باشید. شما باید بر اساس انتخاب های محدود، بهترین انتخاب خود را بکنید.

همانطور که گفتیم این راه حل لزوما بهترین جواب ممکن را به شما نمی‌دهد، زیرا آخر عاقبت طمع لزوما خوش نیست:)
