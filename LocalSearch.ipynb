این مجموعه شامل اعداد $0$ تا $n$ می‌باشد (خود $n$ را شامل نیست.). این مجموعه به ترتیب شماره با ارزش ترین کالا ها را نمایش می‌دهد.

حالا شما باید راه حل حریصانه خود را پیاده سازی کنید. شما تابعی دارید که به ترتیب $n$ ،  $k$ ، $V$ و $W$ را دریافت می‌کند.

خروجی تابع شما لیستی از یک ها و صفر هاست، که عدد $i$ ام به موجودیت کالای $i$ اشاره دارد.

این تابع را به عنوان نمونه ما تکمیل می‌کنیم اما بدانید که جلوتر باید دست به کد شوید!
    </font>
</div>

```python
def do_greedy_by_values(n, k, V, W):
    # this function does greedy by values
    
    answer = [0] * n
    current_capacity = k
    
    I = sort_by_values(n, V)
    
    for i in I:
        if current_capacity >= W[i]:
            answer[i] = 1
            current_capacity -= W[i]
    
    return answer
```

خیلی اوقات ما بدون این‌که بدانیم از این مدل سازی ها برای راحت‌تر شدن کار خود استفاده می‌کنیم.

این مدل‌سازی ها به ما کمک می‌کند تا متوجه شویم که دقیقا چه می‌خواهیم. این اتفاق ما را به رسیدن به جواب نزدیک‌تر می‌کند.
        
حال بیایید تا مساله خود را مدل‌سازی کنیم!)

در این مساله ما $n$ کالا  (شی) داریم که هر کالا ارزش $v_i$ را دارد. می‌دانیم که هر کالا در کوله‌پشتی وجود دارد یا وجود ندارد که می‌توانیم این ویژگی را با $x_i$ به نمایش بگذاریم. این مقدار همواره $0$ یا ‌$1$ می‌باشد. این امر را می‌توانیم به شکل زیر بیان کنیم:
$$x_{i}\in \{0,1\}$$

می‌دانیم که ارزش ذاتی و وزن ذاتی هر کالا، همان وزن و ارزش آن است که به خود کالا برمی‌گردد. انتخاب کردن یا نکردن آن کالا تاثیری در ارزش و وزن ذاتی آن ندارد.


حالا بیایید تا مفهوم ارزش حقیقی و وزن حقیقی هر کالا را برای خود تعریف کنیم.
ما می‌دانیم که اگر کالایی در کوله‌پشتی نباشد، ارزش و وزنی برای ما نخواهد داشت. در نتیجه نه هزینه (وزن) و نه سود (ارزش) ای از آن ما را شامل می‌شود.
ارزش حقیقی و وزن حقیقی کالاهایی که در کوله‌پشتی نیستند را می‌توانیم $0$ در نظر بگیریم.

می‌دانیم ارزش حقیقی ای که هر کالا برای ما دارد $x_i \times v_i$ می‌باشد. چرا؟ به این علت که اگر کالا در کوله‌پشتی باشد $x_i$ برابر با $1$ است در نتیجه  $x_i \times v_i$ برابر $v_i$ می‌شود. پس این کالا برای ما $v_i$ ارزش دارد. حالا اگر این کالا در کوله‌پشتی نباشد، $x_i$ برابر با $0$ است در نتیجه  $x_i \times v_i$ برابر $0$ می‌شود. پس این کالا برای ما هیچ ارزشی ندارد. مهم نیست که ارزش ذاتی کالا چقدر است، اگر کالا درون کوله‌پشتی نباشد هیچ ارزشی برای ما ندارد (ارزش حقیقی آن $0$ است.).

با این اوصاف می‌توانیم بیان کنیم که ارزش کوله‌پشتی جمع‌آوری شده، برابر مجموع ارزش های حقیقی است. زیر در نظر گرفتیم که اگر کالا در کوله‌پشتی نباشد ارزش ذاتی آن بی‌اهمیت است و ارزش حقیقی آن $0$ می‌باشد. اعلام کردیم که اگر کالا درون کوله‌پشتی باشد ارزش حقیقی آن $v_i$ می‌شود.

همین موضوع برای وزن نیز صدق می‌کند. وزن کوله‌پشتی جمع‌آوری شده توسط ما، بنا به استدلال قبلی این بار برای وزن،‌برابر مجموع وزن حقیقی تمام کالا هاست.

در نتیجه می‌توانیم وزن و ارزش کوله‌پشتی جمع‌آوری شده را به صورت زیر بنویسیم:

$$ \text{Total value}= \sum_{i=1}^{n}v_i \times x_i$$
$$ \text{Total weight} = \sum_{i=1}^{n}w_i \times x_i$$

هدفی که ما داریم بیشینه کردن ارزش کوله‌پشتی است. همزمان باید حواسمان به وزن کوله‌پشتی جمع آوری شده باشد تا از محدودیت گذر نکند. می‌توانیم مدل مساله را به صورت زیر بنویسیم:
        
$$ \text{maximize}  \sum_{i=1}^{n}v_i \times x_i  \ \ \ \ \ \ \text{s.t} \ \ \ \ \ \ \  \sum_{i=1}^{n}w_i \times x_i \leqslant k$$
$$x_{i}\in \{0,1\}$$

این عبارت بیانگر این مورد است که ما می‌خواهیم ارزش کوله‌پشتی را بیشینه کنیم به نحوی که ($\text{s.t}$) وزن کوله‌پشتی جمع‌شده از محدودیت وزن کوله‌پشتی گذر نکند.

ما به دنبال $X$ ای هستیم که $\lvert X \rvert = n$ (تعداد اعضای آن برابر n باشد.) و $x_i \in X \Rightarrow x_i \in \{0,1\}$ (همه اعضای آن صفر یا یک باشند.).

$X$ همان جواب مساله ما و خروجی تابع ما می‌باشد.

ساده‌ترین ایده‌ای که برای حل این مساله می‌تواند به ذهن همه برسد، بررسی همه حالات است. در این راه حل ما همه جواب های ممکن را بررسی می‌کنیم و بهترین جواب را خروجی می‌دهیم.

این راه حل مانند روش حریصانه نخواهد بود و به وضوح بهترین جواب را به ما خروجی خواهد داد.

می‌دانیم به ازای هر $X$ با شرط‌های ذکر شده، یک جواب برای مساله خود داریم. ما باید تمام $X$ های موجود را بررسی کنیم و ببینیم که کدام موردِ ممکن،  دارای بیشترین ارزش است.

می‌دانیم که $X$ دارای $n$ عضو است که هر عضو آن را $0$ یا $1$ تشکیل می‌دهد. در نتیجه هر عضو $X$ دارای دو حالت است پس (بنا به اصل ضرب!) ما $2^n$ مقدار $X$ متمایز داریم.

حال برای بررسی همه این حالات باید ایده‌ای داشته باشیم. باید بتوانیم حالت‌های جدید را تولید کنیم.

یکی از ایده های ساده‌ای که وجود دارد، کمک گرفتن از اعداد در مبنای دو است. اگر جواب $X$ را داشته باشیم که $X = \{x_1, x_2, x_3, \dots\, x_n\}$ می‌توانیم عدد $B$ را این‌گونه تعریف کنیم:
$$B = x_1 x_2 x_3 \dots x_n$$
عدد $B$ شامل ارقام 0 و 1 است، در نتیجه این عدد در مبنای دو می‌باشد.

می‌توانیم نتیجه بگیریم که به ازای هر $X$ یک عدد $B$ که در مبنای دو است وجود دارد.

می‌دانیم که $B$ دارای $n$ رقم است، در نتیجه در تناظر هر $B$ (عدد $n$ رقمی در مبنای دو) یک عدد صحیح $q$ بین 0 تا $2^n$ وجود دارد.

چرا این گزاره صحیح است؟ زیرا می‌توانیم $B$ را به مبنای ده ببریم. اگر $B$ را به مبنای ده ببریم می‌تواند حداقل 0 و حداکثر $ 2^n - 1$ باشد.


کافیست اعداد صحیح بین  $0$ و $2 ^ n$ را به مبنای دو ببریم تا هر یک بیانگر یک حالت جواب($X$) باشد.

یا به عبارتی روند زیر را برای رسیدن به یک جواب $X$ طی کنیم:
$$ q \ \ \ \rightarrow \ \ \ B \ \ \ \rightarrow \ \ \ X$$ 
        
به طور مثال نمونه زیر برای $n=3$ یا همان زمانی است که $3$ کالا داریم:
$$
0  \rightarrow \ \ \   000   \rightarrow \ \ \  \{0, 0, 0\} \\
1  \rightarrow \ \ \   001   \rightarrow \ \ \  \{0, 0, 1\} \\
2  \rightarrow \ \ \   010   \rightarrow \ \ \  \{0, 1, 0\} \\
3  \rightarrow \ \ \   011   \rightarrow \ \ \  \{0, 1, 1\} \\
4  \rightarrow \ \ \   100   \rightarrow \ \ \  \{1, 0, 0\} \\
5  \rightarrow \ \ \   101   \rightarrow \ \ \  \{1, 0, 1\} \\
6  \rightarrow \ \ \   110   \rightarrow \ \ \  \{1, 1, 0\} \\
7  \rightarrow \ \ \   111   \rightarrow \ \ \  \{1, 1, 1\} 
$$
با توجه به این توضیحات، تابعی برای حل سوال با بررسی همه حالات می‌نویسیم:
    </font>
</div>

```python
def do_bruteforce(n, k, V, W):
    # Do bruteforce
    
    # initialize answer
    max_total_value = 0
    result = []
    
    # check for i(q) in [0, 2^n)
    for i in range(2 ** n):
        X = []
        X_length = 0
        
        # base 10 to binary
        while i != 0:
            x = i % 2
            i //= 2
            X.append(x)
            X_length += 1
            
        X += [0] * (n - X_length)
        X.reverse()  ###  Extra Computation But make it more clear
        
        #print(X)  # view X 
        
        # Evaluate our answer(X)
        total_value = 0
        total_weight = 0
        
        for j in range(n):
            total_weight += W[j]*X[j]
            total_value += V[j]*X[j]
        
        # Check if we find better answer(X)
        if total_value > max_total_value and total_weight <= k:
            max_total_value = total_value
            result = X[:]
    
    return result
```
